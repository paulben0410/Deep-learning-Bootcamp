```python
import pandas as pd
df = pd.read_excel("data.xlsx")
```

```python
df.rename(columns=df.iloc[0], inplace=True)
df = df.drop(df.index[0])
df = df.dropna(subset=['label'])
df = df.dropna(axis=1)
df = df.drop(columns=['번호','날짜','시간','시편상태','요약','수막두께'])
df = df.apply(pd.to_numeric)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
X = df.drop(['label'], axis=1).values
y = df['label'].values
```

```python
X = StandardScaler().fit_transform(X)
```

```python
df.columns
```

```python
features = ['외부 대기온도', '최대 대기온도', '최소 대기온도', '외부 상대습도', '이슬점 온도', '풍속', '풍정',
       '최대풍속', '체감온도', '열지수', 'THW', 'THSW', '기압', '강우량', '강우강도', 'Rad.',
       'Energy', 'Rad. ', 'UV ', 'UV Dose', 'Hi UV', 'Head D-D', 'Cool D-D',
       'In Temp', 'In Hum', 'In Dew', 'In Heat', 'In EMC', 'In Air Density',
       'ET ', '노면 온도1', 'Wind Samp', 'Wind Tx', 'ISS Recept', 'Arc. Int.']
```

```python
# 표준화된 데이터
pd.DataFrame(X, columns=features)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data=principalComponents, columns = ['1','2'])
```

```python
principalDf.head()
```

```python
import matplotlib.pyplot as plt
```

```python
x_pca = []
y_v = []
y_va = []
y_var = []

for i in range(1,36):
    x_pca.append(i)
    pca = PCA(n_components=i)
    principalComponents = pca.fit_transform(X)
    principalDf = pd.DataFrame(data=principalComponents)
    y_var.append(sum(pca.explained_variance_ratio_))
    y_va.append(pca.explained_variance_ratio_[i-1])
    y_v.append(pca.explained_variance_[i-1])
```

```python
fig = plt.figure(figsize=(20,8))
plt.plot(x_pca,y_var, marker='o', markersize=15, linewidth=4, markeredgecolor='w', label='Cumulative')
plt.bar(x_pca, y_va, color='orange', label='Individual')
plt.grid(True)
plt.xlabel('PCA', fontsize=15)
plt.ylabel('Explained Variance Ratio', fontsize=15)

for i, v in enumerate(x_pca):
    plt.text(v, y_var[i], round(y_var[i],2),                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 14, 
             horizontalalignment='right',  # horizontalalignment (left, center, right)
             verticalalignment='baseline')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(20,8))
plt.plot(x_pca, y_v, marker='o', markersize=15, linewidth=4, markeredgecolor='w')
plt.grid(True)
plt.xlabel('PCA', fontsize=15)
plt.ylabel('Explained Variance', fontsize=15)
```
